# Part of Speech Tagging: an example
## What is Part-of-Speech (POS) Tagging? 

**Part-of-Speech tagging (POS)** is a Natural Language Processing (NLP) technique that assigns grammatical labels to each word in a text. POS tagging analysis can help us analyze sentence structures, identify patterns in literary language, compare writing styles, support higher-level tasks like parsing or sentiment analysis.

As an example, we can analize a sentence from the play:

Tis but thy name that is my enemy.
Thou art thyself, though not a Montague.
What’s Montague? It is nor hand, nor foot,
Nor arm, nor face. O, be some other name
Belonging to a man.
What’s in a name? That which we call a rose

In a basic way we can devide the text into:

| Family | What They Do | Examples |
|------|-------------|----------|
| Nouns | Name things or ideas | *name, rose, Romeo* |
| Verbs | Show actions or states | *love, be, call* |
| Adjectives | Describe nouns | *fair, gentle* |
| Adverbs | Describe actions or qualities | *softly, truly* |
| Function words | Hold sentences together | *the, and, of* |

### Libraries

POS tagging can be done:
- **by humans** (grammar)
- **by annotation projects** (ex. MorphAdorner)
- **by software libraries** (automatic taggers, ex NLTK)

## 2. NLTK (Natural Language Toolkit)
NLTK is a natural language processing (NLP) library that can be used for a variety of tasks such as tokenization, stemming, lemmatization, and in this case, Part of Speech Tagging.

NLTK’s default POS tagger is **statistical, pre-trained model**, meaning that it works by learning patterns from previously annotated texts and use **probability** to assign a tag to each token, considering not only the token but its context. The tags used in the NLTK POS tagger are from the **[Penn Treebank POS tagset](https://www.ling.upenn.edu/courses/Fall_2003/ling001/penn_treebank_pos.html)**, which is a **standardized set of labels** used to describe the grammatical role (part of speech) of each word in a text (ex. NN for nouns, VBZ for verbs, CC for coordinating conjunctions such as *and*, *but*, *or* etc.)

Important to note: it is trained on modern English and it does not understand historical grammar (could be a limitation for older text). 

````python
import nltk
````

````python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

from nltk import word_tokenize, pos_tag

# Text from Romeo and Juliet
text = """'Tis but thy name that is my enemy.
Thou art thyself, though not a Montague.
What's Montague? It is nor hand, nor foot,
Nor arm, nor face. O, be some other name
Belonging to a man.
What's in a name? That which we call a rose"""

# Tokenize the text
tokens = word_tokenize(text)

# Apply POS tagging
pos_tags = pos_tag(tokens)

# Display results
for word, tag in pos_tags:
    print(f"{word}: {tag}")
````

````python
# 1. Import libraries
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk import pos_tag
from collections import Counter
import requests
````

````python
# Download required NLTK data
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
````

````python
# 2. Download the Romeo and Juliet text from the web
url = "https://folger-main-site-assets.s3.amazonaws.com/uploads/2022/11/romeo-and-juliet_TXT_FolgerShakespeare.txt"
response = requests.get(url)
text = response.text

print(text[:500])
````

````python
# 3. Tokenize into sentences
sentences = sent_tokenize(text)

# 4. Tokenize each sentence into words
tokenized_sentences = [word_tokenize(sent) for sent in sentences]

# 5. Apply POS tagging
tagged_sentences = [pos_tag(sent) for sent in tokenized_sentences]
````

````python
# 6. Flatten the list of tagged words
tags = [tag for sent in tagged_sentences for tag in sent]

# 7. Show first 20 tagged words
print(tags[:20])
````

````python
# 8. Count the POS tags
pos_counts = Counter(tag for word, tag in tags)
print("Most common POS tags:", pos_counts.most_common(10))
````

````python
# 9. Optional: Group into major categories
noun_tags_nltk = ['NN', 'NNS', 'NNP', 'NNPS']
verb_tags_nltk = ['VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ']
adj_tags_nltk = ['JJ', 'JJR', 'JJS']
adv_tags_nltk = ['RB', 'RBR', 'RBS']

noun_count_nltk = sum(1 for word, tag in tags if tag in noun_tags_nltk)
verb_count_nltk = sum(1 for word, tag in tags if tag in verb_tags_nltk)
adj_count_nltk = sum(1 for word, tag in tags if tag in adj_tags_nltk)
adv_count_nltk = sum(1 for word, tag in tags if tag in adv_tags_nltk)

print("Counts - Nouns:", noun_count_nltk, "Verbs:", verb_count_nltk, "Adjectives:", adj_count_nltk, "Adverbs:", adv_count_nltk)
````

### MorphAdorner
In the previous method, we saw how POS analysis can be done on raw text (a .txt file, in this case) using a pre-trained model that automatically decides which tag to assign to each token based on its previous knowledge (the NLTK library). However, we also noted that, because such models are trained mostly on modern English text, some Early Modern English expressions (like those in Shakespeare) might be more difficult to classify correctly.

The Folger Library provides some texts [pre-annotated with POS tags using **MorphAdorner** ](https://www.folger.edu/explore/shakespeares-works/download/#romeo-and-juliet), which is better suited for Early Modern English because it relies on grammatical rules and dictionary definitions of words, rather than purely statistical patterns. Unlike NLTK, MorphAdorner does not use Penn Treebank tags; instead, it uses a custom tagset designed for Early Modern English, which is why it works better for Shakespeare. Using these pre-annotated texts, we can directly extract POS information for analysis.

The text comes in XML format, where tokenization is already done, and lemmas and POS tags are already assigned.

Tokens in the XML appear in this format:

`<w xml:id="fs-rom-0000080" n="PRO.1" lemma="two" ana="#crd">Two</w>`

in which:

```
<w> indicates the precence of a word token
Two	indicates the "surface form" (meaning what appears in the text)
lemma="two"	indicates the dictionary/base form
ana="#crd"	indicates the Part-of-speech category
xml:id	is the unique identifier
```
Using the following code, we will extract the POS tags found in the text.

````python
#importing the needed libraries
from lxml import etree
from collections import Counter
````

We have previously downloaded the TEISimple XML (annotated with MorphAdorner for part-of-speech analysis) version of the "*Romeo and Juliet*" play found at this [link](https://www.folger.edu/explore/shakespeares-works/download/#romeo-and-juliet).
````python
import zipfile

zip_file = "data/romeo-and-juliet_TEIsimple_FolgerShakespeare.xml_.zip" #path of the zip file, starting from the current directory
extract_folder = "data/romeo_and_juliet_tei" #path of the folder where the extracted material will go to 

with zipfile.ZipFile(zip_file, "r") as z:
    z.extractall(extract_folder) #command to extract the material and put it in the extract folder

import os
os.listdir("data/romeo_and_juliet_tei")
````

````python
# Path to your TEI XML file
xml_path = "data/romeo_and_juliet_tei/romeo-and-juliet_TEIsimple_FolgerShakespeare.xml"

# Parse XML
tree = etree.parse(xml_path)
root = tree.getroot()
````

````python
ns = {"tei": "http://www.tei-c.org/ns/1.0"}
````

````python
# Find all word elements
words = root.findall(".//tei:w", namespaces=ns)

len(words)
````

````python
pos_tags = []

for w in words:
    pos = w.get("ana")
    if pos is not None:
        pos_tags.append(pos)
````

````python
# Count frequencies
pos_freq = Counter(pos_tags)

# View the 20 most common POS tags
print(pos_freq.most_common(20))
````

This results show us a high-frequency of nouns (#n1, #n2, #n3) reflecting the narrative nature of the text that has frequent names of characters, places and objects. We can also see the richness of the analysis, as the tags can distinguish verbs tenses (ex. #vvb → base form verbs, #vvi → infinitive or other forms, #vvz → present tense 3rd person singular, #vvd → past tense, #vmb → modal verbs, #vvn → past participles).

For readability and to allow comparison with Penn Treebank–based analyses, MorphAdorner POS tags were grouped into broad grammatical categories (nouns, verbs, adjectives, and adverbs) according to their tag prefixes. However, this grouping inevitably loses some of the detailed information provided by MorphAdorner, so this step is optional and not recommended for more in-depth research.

````python
# Group POS tags into major grammatical categories

noun_count = sum(1 for tag in pos_tags if tag.startswith("#n"))
verb_count = sum(1 for tag in pos_tags if tag.startswith("#v"))
adj_count  = sum(1 for tag in pos_tags if tag.startswith("#j"))
adv_count  = sum(1 for tag in pos_tags if tag.startswith("#av"))

print(
    "Counts -",
    "Nouns:", noun_count,
    "Verbs:", verb_count,
    "Adjectives:", adj_count,
    "Adverbs:", adv_count
)
````

````python
import matplotlib.pyplot as plt
import numpy as np
````

## Comparison between POS tagging: a Visualization

````python
# POS categories
categories = ["Nouns", "Verbs", "Adjectives", "Adverbs"]

# MorphAdorner counts
morph_counts = [
    noun_count,
    verb_count,
    adj_count,
    adv_count
]

# NLTK counts
nltk_counts = [
    noun_count_nltk,
    verb_count_nltk,
    adj_count_nltk,
    adv_count_nltk
]
````

````python
x = np.arange(len(categories))
width = 0.35

plt.figure(figsize=(10, 6))

plt.bar(x - width/2, morph_counts, width, label="MorphAdorner")
plt.bar(x + width/2, nltk_counts, width, label="NLTK")

plt.xticks(x, categories)
plt.ylabel("Frequency")
plt.title("POS Category Comparison: MorphAdorner vs NLTK")
plt.legend()

plt.tight_layout()
plt.show()
````

This chart compares the distribution of major part-of-speech categories as identified by MorphAdorner and NLTK. While both methods show similar overall trends, differences in category frequencies reflect the distinct tagging strategies and linguistic assumptions underlying each system.

Although both taggers identify similar grammatical patterns, MorphAdorner’s design for Early Modern English leads to systematic differences when compared with NLTK’s Penn Treebank–based tagging.
